```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.linspace(0,10,100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color = "red",linestyle = "--",label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("machine learning")
plt.axis([-1,11,-2,2])
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color="red")
```
